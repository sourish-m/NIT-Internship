```python
import pandas as pd
import numpy as np
import os
```

```python
dataset = pd.read_csv('final_filtered_checked_fin_flood_normal.csv')
dataset = dataset.drop(['Unnamed: 0'], axis = 1)
dataset = dataset [['antecedents','consequents']]
dataset
```

```python
#Created packet array p[]

filename = 'Monday_Normal202_Cfile_output.txt'
with open(filename, 'r') as file:
    p = file.read().splitlines()
# print(p)
```

```python
print(len(p)) # length of packet array
```

```python
def string_to_list(string):
  #string.strip()
  substrings = string.split(',')
  new_list = []
  for substring in substrings:
    new_list.append(substring.strip())
  return new_list
```

```python
#Creation of Signature Array sign[]

sign = []

for row in dataset.itertuples():
  combine = string_to_list(row[1]) + string_to_list(row[2])
  sign.append(combine)

print(sign)
```

```python
print(len(sign)) # length of signature array
```

```python
def checker(list1, packet):
  for ele in list1:
    if ele in packet:
      continue
    else:
      return 0
  return 1
```

```python
j = 0
i = 0
d = [0] * len(p)
while j < len(p):
  i = 0
  while i < len(sign):
    if checker(sign[i], p[j]):
      d[j] += 1

    i += 1

  j += 1
```
