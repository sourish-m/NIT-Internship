```python

import pandas as pd

dataset = pd.read_csv('unfiltered check_output for fin_flood 8-4.csv')
dataset = dataset.drop(['Unnamed: 0'], axis = 1)
dataset['Frequency'] = dataset['AttackFreq']

tempList  = []
tempList2 = []
row  = 0
row2 = 0
num = len(dataset)
while row < num:
    tempList = dataset.loc[row,'antecedents'] + dataset.loc[row, 'consequents']
    row2 = row + 1
    print(num)
    while row2 < num:
        tempList2 = dataset.loc[row2,'antecedents'] + dataset.loc[row2, 'consequents']

        if set(tempList) == set(tempList2):
            dataset.at[row, 'Frequency']+=dataset.at[row2, 'AttackFreq']
            # print(row, row2)
            dataset.drop(row2, inplace=True)
            dataset = dataset.reset_index(drop=True)


        num = len(dataset)
        row2 +=1

    row +=1


print(dataset)

# dataset.to_csv("filtered.csv")
```
