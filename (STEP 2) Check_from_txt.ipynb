# **Checking for Normal Data**

```python
def string_to_list(string):
  #string.strip()
  substrings = string.split(',')
  new_list = []
  for substring in substrings:
    new_list.append(substring.strip())
  return new_list
```

```python
def checker(list1, packet):
  for ele in list1:
    if ele in packet:
      continue
    else:
      return 0
  return 1
```

```python
sign = []

for row in dataset.itertuples():
  combine = string_to_list(row[1]) + string_to_list(row[2])
  sign.append(combine)

print(sign)
```

```python
normalFreq = [0] * len(dataset)
for line in normalData:
  i = 0
  for lst in sign:
    if checker(lst, line):
      normalFreq[i]+=1
    i+=1

print(normalFreq)
```
